```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def signal(x):
    if x < np.pi and x > 0:
        return np.sin(x)
    else:
        return 0
```

```python
x = np.linspace(0, 2*np.pi, 100)
y = [signal(item) for item in x]
plt.plot(x, y)
plt.show()
```
